Chia doi danh sach de tim kiem (mid, start, end)
**Do phuc tap: O(log2(n))**

```python
def binary_search_loop(arr, target) -> int:
    #danh sach rong
    if not arr: return -1
    #tao mid, start, end
    start, end = 0, len(arr) -1
    mid = (end - start) // 2 + start
    while start <= end:
        if arr[mid] == target:
            return mid
        if arr[mid] < target:
            start = mid + 1
        else:
            end = mid - 1
        mid = (end - start) // 2 + start
    return -1
```

```python
import timeit
import random
arr = [random.randint(-10, 100) for _ in range(1000)]
arr.sort()
arr
```

```python
print("Do thoi gian tim kiem: ", timeit.timeit(lambda: binary_search_loop(arr, -30), number = 1000))
```

```python
def binary_search_recursive(arr, target, start, end):
    if start > end: return -1
    mid = (end - start)//2 + start
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, target, mid+ 1, end)
    else:
        return binary_search_recursive(arr, target, start, mid - 1)
```

```python
def run(arr, target):
    if not arr:return 1
    return binary_search_recursive(arr, target, 0, len(arr)- 1)
```

```python
import timeit
import random
arr2 = [ random.randint(-10, 100) for _ in range(10000)]
arr2.sort() 
# ham lambda 
def a(): return run(arr2, random.randint(-100, 1000))
exe_time = timeit.timeit(a, number=1000)
print(f"Ket qua tim kiem: {a()}")
print("Do thoi gian tim kiem: ", exe_time)
```
